# Popularity Based Recommender System

```python
# Merging the ratings dataset with the books dataset on the 'ISBN' column  
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.head(1)
```

```python
# Grouping by 'Book-Title' to count the number of ratings each book received
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
# Renaming the 'Book-Rating' column to 'num_ratings' for better readability
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
# Grouping by 'Book-Title' to calculate the average rating for each book  
avg_rating_df = ratings_with_name.groupby('Book-Title').mean(numeric_only=True)['Book-Rating'].reset_index()
# Renaming the 'Book-Rating' column to 'avg_rating' for better readability  
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
# Merging the number of ratings dataframe with the average ratings dataframe on 'Book-Title'  
# This combines both the number of ratings and the average rating for each book  
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_rating',ascending=False).head(50)
```

```python
popular_df
```

```python
# Book author name and image, publisher name
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_rating']]
```

```python
popular_df
```

# Collaborative Filtering Based Recommender System

```python
# then Experiance user rating>200
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
x[x]
```

```python
padhe_likhe_users = x[x].index
```

```python
# Filtering the ratings dataset to include only ratings from users who have rated more than 200 books 
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
filtered_rating.head()
```

```python
# Grouping by 'Book-Title' to count the number of ratings for each book  
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50

# Filtering books that have received at least 50 ratings  
y[y]
```

```python
# Extracting the titles of books that have received at least 50 ratings 
famous_books = y[y].index
famous_books
```

```python
# Filtering the ratings dataset to include only ratings for the most rated books 
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# Creating a pivot table where rows represent book titles, columns represent users,  
# and values represent the ratings given by users to books  
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
# Filling missing values (NaN) with 0 to indicate that a user hasn't rated a book  
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Computing the cosine similarity between books using the pivot table  
similarity_scores = cosine_similarity(pt)
```

```python
cosine_similarity(pt).shape
```

```python
def recommend(book_name):
    # Fetching the index of the given book from the pivot table
    index = np.where(pt.index == book_name)[0][0]

    # Calculating similarity scores and sorting them in descending order
    # Skipping the first one as it will be the book itself
    similar_items = sorted(list(enumerate(similarity_scores[index])), key=lambda x: x[1], reverse=True)[1:5]

    data = []
    for i in similar_items:
        item = []
        # Fetching book details from the books dataset based on the matched book title
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]

        # Extracting book title, author, and image URL, ensuring unique values
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        # Appending the book details to the recommendations list
        data.append(item)

    return data
```

```python
recommend('1984')
```
